```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio: tamaño del dominio y malla
Lx, Ly = 1.0, 1.0  # longitud en x e y
Nx, Ny = 100, 100  # resolución

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)

# Frecuencias que quieres visualizar
m, n = 2, 2  # puedes cambiar estos valores para otras bases

# Función base compleja 2D
phi_mn = np.exp(2j * np.pi * (m * X / Lx + n * Y / Ly))

# Visualización
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].imshow(np.real(phi_mn), extent=(0, Lx, 0, Ly), origin='lower', cmap='viridis')
axs[0].set_title(f'Real(φ_{m},{n})')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].imshow(np.imag(phi_mn), extent=(0, Lx, 0, Ly), origin='lower', cmap='plasma')
axs[1].set_title(f'Imag(φ_{m},{n})')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.tight_layout()
plt.show()
```
